# 9th Tutorial: Cleaning Data - Casting Datatypes and Handling Missing Values

```python
import numpy as np
import pandas as pd
```

```python
people ={
    'first' : ['Corey' , 'Jane' , 'John' , 'Chris' , np.nan , None , 'NA'],
    'last'  : ['Schafer' , 'Doe' , 'Doe' , 'Schafer' , np.nan , np.nan , 'Missing'],
    'email' : ['CoreySchafer@gmail.com' , 'JaneDoe@gmail.com' , 'JohnDoe@gmail.com' , None , np.nan , 'Anonomous@email.com' , 'NA' ],
    'age'   : [ '33' , '55' , '63' , '36' , None , None , 'Missing']
}
```

```python
df = pd.DataFrame(people)

df.replace('NA' , np.nan , inplace=True)
df.replace("Missing", np.nan , inplace=True)
```

```python
df
```

```python
df.dropna()
```

```python
df.dropna(axis='index' , how='any')
```

```python
df.dropna(axis='index' , how='all')
```

```python
df.dropna(axis='columns' , how='all')
```

```python
df.dropna(axis='columns')
```

```python
df.dropna(axis='columns' , how='any')
```

```python
df.dropna( subset='email' )
```

```python
df.dropna( axis='index', how='any' , subset='email' )
```

```python
df.dropna( axis='index', how='all' , subset=['email' , 'last'] ) 
# To save the DataFrame use the inplace=True argument
```

```python
df.fillna(0) # To save the DataFrame use the inplace=True argument
```
