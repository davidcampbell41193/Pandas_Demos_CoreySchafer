## Dictionary to Panda DataFrame

```python
# Define a Python dictionary
people ={
    'first' : ['David' , 'Sam' , 'Justin'],
    'last'  : ['Campbell' , 'Fingerman' , 'Jozokas'],
    'email' : ['davidcampbell@gmail.com' , 'samfingerman@gmail.com' , 'justinjozokas@gmail.com']
}

# Turn the Dictionary into a Pandas DataFrame
ex_people = pd.DataFrame(people)
ex_people
```

We can access the 'email' as a key the dictionary.

```python

ex_people['email']
```

We can also access the 'email' from the DataFrame class as an attribute. 

If there is column with the same name as an attribute associated with the DataFrame, then accessing the column as attribute will cause an error. In this case, the column must be accessed at the key to a dictionary.

```python
ex_people.email
```

 Multiple collumns of a DataFrame can be accessed by passing a list of keys.

```python

ex_people[['last', 'email']]
```

```python
# Access multiple Columns and Rows
# Note the the order in the Display is the same the ordering the DataFrame.loc[] function
ex_people.loc[ [1,0] , [ 'email' , 'last'] ]
```

Columns is an attribute of the DataFrame class the return an Index Pandas class. The Index class contains a list of the the different columns names.

```python
ex_people.columns
```

```python
type(ex_people.columns)
```

Shape is an attribute of the DataFrame, Series, and Index classes in Pandas.

```python
print( df.columns.shape , ex_people.columns.shape )
```

```python
# Pandas suports two classes: DataFrame and Series
type(ex_people)
```

```python
# If we access one column of a DataFrame its class is Series
type(ex_people.email)
```

# 3rd Tutorial: Indexes - How to Set, Reset, and Use Indexes

```python
# Define a Python dictionary
people ={
    'first' : ['David' , 'Sam' , 'Justin'],
    'last'  : ['Campbell' , 'Fingerman' , 'Jozokas'],
    'email' : ['davidcampbell@gmail.com' , 'samfingerman@gmail.com' , 'justinjozokas@gmail.com']
}

# Turn the Dictionary into a Pandas DataFrame
ex_people = pd.DataFrame(people)
ex_people
```

The set_index method can be used to the index of a column of the DataFrame.

```python
ex_people.set_index('email' , inplace = True )

# pd.DataFrame.set_index( ex_people , 'email' , inplace=True )
```

```python
ex_people
```

.index is another attribute of the Pandas DataFrame class. It returns an object with the Pandas class Index.

```python
ex_people.index
```

We can now access a row by their email.

```python
ex_people.loc['davidcampbell@gmail.com']
```

```python
#ex_people.reset_index(inplace=True)

pd.DataFrame.reset_index( ex_people , inplace=True )
```

```python
ex_people
```
