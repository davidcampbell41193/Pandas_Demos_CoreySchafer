# Combining Data Sets

```python
#Partition the data frame into a DataFram without the "Blockchain" and a Series with "Blockchain."

df_wo_bs = df.loc[:, df.columns != 'Blockchain']
df_w_bs = df['Blockchain']

print( 'Full Data Frame, Smaller Data Frame, Blockchain Series'  )
print( df.shape , '\t' , df_wo_bs.shape, '\t' , df_w_bs.shape )
```

What does inplace do? Are the shapes the same in each? Are the amount of rows?

Inplace save the chagnes the the dropna function does. No, the shapes are not the same in each. The dropna reduces the number of entries/row in the ps.Series. 

Comparing the Blockchain (BC) shape after the before and after .dropna() function we see the number of rows decreases.

```python
df_w_bs.dropna(inplace=True)


print( 'BC Series without NaN' )
print( df_w_bs.shape )
```

Merge `df_wo_bs` with `df_w_bs` to form the original dataframe.

Verify they are the same.

```python
# Insert the Blockchain Series into the DataFrame without Blockchain at position 42.

df_wo_bs.insert(42 , 'Blockchain' , df_w_bs )

# Rename the DataFrame

df_prime = df_wo_bs
```

```python
# Confirm the that "primed" DataFrame has the same shape of the original DataFrame

print( df_prime.shape )

df_prime.shape == df.shape
```

```python
#Verify that the Blockchain columns was put in corrct place (ie. Position 42).

df_prime.columns.get_loc('Blockchain') 
```

```python
# Becase NaN == NaN returns false I a going to replace NaN with "NA"

df = df.fillna("NA")
df_prime = df_prime.fillna("NA")
```

```python
df == df_prime
```

```python
# The dataframes seem to match. Let us confirm

x = (df == df_prime)

y = x.sum()

y
```

```python
#If equal column is identically true then the sum should equal the total number of rows, 73268.

print( 'The total number of rows in df are:\n',df.shape[0] )

z = (y == df.shape[0])

z
```

```python
"""
If each colums is identically true, is true, then the sum over z should equal the number
of row columns in the original DataFrame, df.
""" 

w = z.sum()

print( df.shape[1] )

w == df.shape[1]
```
